```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../assets/times-forest.csv")
```

```python
df_no_ids = df.query("ALG != 'IDS'")
```

```python
# Plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a lineplot for each ALG and Case
# sns.lineplot(data=df.query("ALG == 'IDS'"), x="Distance", y="Time",hue="Case", markers=True, dashes=False)
sns.lineplot(data=df_no_ids.query("Case == 'Reverse'"), x="Distance", y="Time", hue="ALG", markers=True, dashes=False)

# Add title and labels
plt.title("Time per Distance for IDS", fontsize=14)
plt.xlabel("Distance")
plt.ylabel("Time")

# Show legend and plot
plt.legend(title="Case")
plt.show()
```
