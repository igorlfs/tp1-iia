# Visited

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../assets/visited-forest.csv")
```

```python
# Plot using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Create a lineplot for each ALG and Case
sns.lineplot(data=df.query("Case == 'Base'"), x="Distance", y="Visited", hue="ALG", markers=True, dashes=False)
# sns.lineplot(data=df.query("ALG == 'IDS'"), x="Distance", y="Time",hue="Case", markers=True, dashes=False)
# sns.lineplot(data=df_no_ids.query("Case != 'Reverse'"), x="Distance", y="Time", hue="ALG", markers=True, dashes=False)

# Add title and labels
plt.title("Nodes visited per Distance - Forest", fontsize=14)
plt.xlabel("Distance")
plt.ylabel("Nodes visited")

# Show legend and plot
plt.legend(title="Case")
plt.show()
```
